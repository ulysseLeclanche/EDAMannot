# Compute the difference in metric values (number of annotations, IC, entropy) between direct annotations and those obtained via enrichment using the subClassOf hierarchy.

```python
import pandas as pd
```

### Loading the dataframes calculated with the init command on ShareFAIR-KG  

```python
dfToolallmetrics = pd.read_csv("../Dataframe/dfToolallmetrics.tsv.bz2", sep="\t") 
#All tool metrics (IC, entropy and number of annotations) on direct and inherited Topics and Operations of tool in bio.tools.
```

```python
dfToolallmetrics_NT = pd.read_csv("../Dataframe/dfToolallmetrics_NT.tsv.bz2", sep="\t")
#All tool metrics (IC, entropy and number of annotations) on direct Topics and Operations of tool in bio.tools.
```

### Explanation of calculated scores 
- topicScore: Score based on the IC (Information Content) of topics associated with a tool.
- operationScore: Score based on the IC of operations associated with a tool.
- score: Sum of the two previous scores.
- topicEntropy: Total entropy of topics associated with a tool.
- operationEntropy: Total entropy of operations associated with a tool.
- entropy: Sum of the previous entropies.

```python
dfToolallmetrics['Heritage'] = "Inherited annotations"
dfToolallmetrics_NT['Heritage'] = "Directly assigned annotations"
```

```python
df_combined = pandas.concat([dfToolallmetrics, dfToolallmetrics_NT], ignore_index=True)
#Define a custom aggregation dictionary for the columns of interest:
agg_funcs = {
'nbTopics': ['min', 'max', 'mean', 'median', 'std',
lambda x: x.quantile(0.25), # 1st quartile
lambda x: x.quantile(0.75)], # 3rd quartile
'nbOperations': ['min', 'max', 'mean', 'median', 'std',
lambda x: x.quantile(0.25), # 1st quartile
lambda x: x.quantile(0.75)] # 3rd quartile
}
# Perform groupby on the Condition column and aggregate:
grouped_stats = df_combined.groupby('Heritage').agg(agg_funcs)
# Rename the lambda columns for clarity:
grouped_stats.columns = ['_'.join(col).strip() for col in grouped_stats.columns.values]
grouped_stats.rename(columns={
'topicEntropy_<lambda_0>': 'topicEntropy_q1',
'topicEntropy_<lambda_1>': 'topicEntropy_q3',
'operationEntropy_<lambda_0>': 'operationEntropy_q1',
'operationEntropy_<lambda_1>': 'operationEntropy_q3'
}, inplace=True)
grouped_stats
```
